```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l[:3]

movie['cast']= movie['cast'].apply(convert)
```

```python
def convert_crew(obj):
    l = []
    for i in ast.literal_eval(obj):
        
        if i['job'] == "Director":
            print
            l.append(i['name'])
    return l

movie['crew'] = movie['crew'].apply(convert_crew)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tags'] = movie['overview']+movie['genres']+movie['cast']+movie['crew']
```

```python
new_df = movie[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_func(x):
    y = []
    for i in x:
       y.append(ps.stem(i))
    return "".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem_func)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
list(enumerate(similarity[4805]))
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 3')
```
